```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

from jupyterthemes import jtplot
jtplot.style(theme="grade3")
```

```python
dataSize = ["256", "512", "1K", "2K", "4K", "8K", "16K", "32K", 
            "64K", "128K","256K", "512K", "1M", "2M", "4M",]

tpt = np.zeros([300, len(dataSize)])
for i in range(len(dataSize)):
    tpt[:, i] = np.loadtxt("./q1/"+"transport_time_"+str(dataSize[i])+"byte.txt", dtype=np.float64) * 1000 # convert "s" to "ms"
```

```python
df = pd.DataFrame(data=tpt, columns=dataSize)

q1fig = plt.figure(figsize=(22,10))
boxplot = df.boxplot(rot=45, fontsize=15, showfliers=False)
boxplot.set_ylabel("Latency [ms]")
boxplot.set_xlabel("Data Size [byte]")
q1fig.savefig("./q1/q1.png", format="png", dpi=600, pad_inches=0)
```
